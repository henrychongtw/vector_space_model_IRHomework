```python
def numDocsContaining(word, doclist):
    doccount = 0
    for d in doclist:
        if word in d:
            doccount += 1 
        
    return doccount

def idf(word, doclist, df):
    n_samples = len(doclist)
    return math.log(n_samples / df,10)
        
def idf_prob(word, doclist, df):
    n_samples = len(doclist)
    return np.log((n_samples - df) / df)

def idf_max(word, df):
    return np.log(1 + ( 2260 / df))

def idf_smooth(word, doclist, df):
    n_samples = len(doclist)
    return np.log(1 + (n_samples / df))

numDocsCount_df_test = {}
my_idf = {}
my_idf_prob = {}
my_idf_smooth = {}
my_idf_max = {}
for word in vocab:
    df_test = numDocsContaining(word, corpus_l2)
    
    numDocsCount_df_test[word] = df_test
    my_idf[word] = idf(word, corpus_l2, df_test)
    my_idf_prob[word] = idf_prob(word, corpus_l2, df_test)
    my_idf_max[word] = idf_max(word, df_test)
    my_idf_smooth[word] = idf_smooth(word, corpus_l2, df_test)
#print(numDocsCount_df)
```
